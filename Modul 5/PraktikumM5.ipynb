```python
# Praktikum 1

# Import library yang diperlukan
import matplotlib.pyplot as plt
import numpy as np

# Fungsi untuk menggambar segitiga
def gambar_segitiga(titik, warna='b', label=None):
    # Tambahkan titik pertama ke akhir untuk menutup segitiga
    segitiga = np.array(titik)
    segitiga = np.vstack([segitiga, segitiga[0,:]])
    plt.plot(segitiga[:, 0], segitiga[:, 1], color=warna, label=label)

# Fungsi untuk melakukan translasi
def translasi(titik, vektor):
    # Menambahkan vektor translasi ke semua titik segitiga
    # titik_baru = [[x + vektor[0], y + vektor[1]] for x, y in titik] # Ini adalah implementasi list comprehension, tapi
    # titik_baru = titik + vektor # Cara yang lebih mudah menggunakan NumPy (asumsi 'titik' adalah np.array)
    
    # Menggunakan implementasi yang terlihat di gambar (list comprehension)
    if isinstance(titik, np.ndarray):
        titik_baru = titik + vektor
    else:
        # Asumsi titik adalah list of lists atau tuple of tuples
        titik_baru = [[x + vektor[0], y + vektor[1]] for x, y in titik]
        
    return titik_baru

# Titik-titik awal segitiga
segitiga_awal = [(1, 2), (3, 5), (6, 2)]

# Vektor translasi
vektor_translasi = (2, 3)

# Translasi segitiga
segitiga_tertranslasi = translasi(segitiga_awal, vektor_translasi)

# Plot segitiga sebelum dan sesudah translasi
plt.figure()

# Gambar segitiga awal
gambar_segitiga(segitiga_awal, warna='b', label='Segitiga Awal')

# Gambar segitiga setelah translasi
gambar_segitiga(segitiga_tertranslasi, warna='r', label='Segitiga Setelah Translasi')

# Menampilkan grid dan setting sumbu
plt.grid(True)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Translasi 2D Segitiga')

# Tampilkan plot
plt.show()
```

```python
# Praktikum 2

# Import library yang diperlukan
import matplotlib.pyplot as plt
import numpy as np

# Fungsi untuk menggambar segitiga
def gambar_segitiga(titik, warna='b', label=None):
    # segitiga = np.array(titik)
    # Tambahkan titik pertama ke akhir untuk menutup segitiga
    segitiga = np.vstack([titik, titik[0,:]]) # Asumsi 'titik' sudah berupa array NumPy
    plt.plot(segitiga[:, 0], segitiga[:, 1], color=warna, label=label)

# Fungsi untuk melakukan translasi
def translasi(titik, vektor):
    # Menambahkan vektor translasi ke semua titik segitiga
    # titik_baru = [[x + vektor[0], y + vektor[1]] for x, y in titik] # Ini adalah implementasi list comprehension yang terlihat di gambar
    # Menggunakan operasi NumPy yang lebih efisien jika 'titik' adalah np.array
    titik_baru = titik + vektor
    return titik_baru

# Fungsi utama untuk input dinamis dan translasi
def main():
    # Input titik segitiga dari pengguna
    print("Input titik segitiga dari pengguna")
    print("Masukkan koordinat segitiga (dalam format x, y):")
    
    # Input Titik 1
    x1, y1 = map(float, input("Titik 1 (x1, y1): ").split(','))
    # Input Titik 2
    x2, y2 = map(float, input("Titik 2 (x2, y2): ").split(','))
    # Input Titik 3
    x3, y3 = map(float, input("Titik 3 (x3, y3): ").split(','))

    # Titik segitiga awal
    segitiga_awal = np.array([[x1, y1], [x2, y2], [x3, y3]])

    # Input vektor translasi dari pengguna
    print("Masukkan vektor translasi (dalam format dx, dy):")
    dx, dy = map(float, input("Vektor translasi (dx, dy): ").split(','))
    vektor_translasi = np.array([dx, dy])

    # Translasi segitiga
    segitiga_tertranslasi = translasi(segitiga_awal, vektor_translasi)

    # Plot segitiga sebelum dan sesudah translasi
    plt.figure()

    # Gambar segitiga awal
    gambar_segitiga(segitiga_awal, warna='b', label='Segitiga Awal')

    # Gambar segitiga setelah translasi
    gambar_segitiga(segitiga_tertranslasi, warna='r', label='Segitiga Setelah Translasi')

    # Menampilkan grid dan setting sumbu
    plt.grid(True)
    
    # Menentukan batas sumbu (dengan padding 10)
    # Gunakan min/max dari semua koordinat titik
    all_x = np.concatenate([segitiga_awal[:, 0], segitiga_tertranslasi[:, 0]])
    all_y = np.concatenate([segitiga_awal[:, 1], segitiga_tertranslasi[:, 1]])
    
    # Kode yang terlihat di gambar (dengan asumsi x1..x3, y1..y3 adalah variabel float)
    # plt.xlim(min(x1, x2, x3) - 10, max(x1, x2, x3) + 10)
    # plt.ylim(min(y1, y2, y3) - 10, max(y1, y2, y3) + 10)
    
    # Implementasi umum dengan min/max dari array NumPy
    plt.xlim(all_x.min() - 3, all_x.max() + 3) # Memberi padding 3
    plt.ylim(all_y.min() - 3, all_y.max() + 3) # Memberi padding 3
    
    plt.gca().set_aspect('equal', adjustable='box')
    plt.legend()
    plt.title('Translasi 2D Segitiga')

    # Tampilkan plot
    plt.show()

# Panggil fungsi utama

main()
```

```python
# Praktikum 4

# Import library yang diperlukan
import matplotlib.pyplot as plt
import numpy as np

# Fungsi untuk menggambar lingkaran
def gambar_lingkaran(tengah, radius, warna='b', label=None):
    # Buat data untuk lingkaran (sudut dari 0 sampai 2*pi)
    theta = np.linspace(0, 2 * np.pi, 100)
    
    # Hitung koordinat x dan y berdasarkan pusat (tengah[0], tengah[1]) dan radius
    x = tengah[0] + radius * np.cos(theta)
    y = tengah[1] + radius * np.sin(theta)
    
    plt.plot(x, y, color=warna, label=label)

# Fungsi untuk melakukan translasi
def translasi(tengah, vektor):
    # Menambahkan vektor translasi ke pusat lingkaran
    # tengah_baru = [tengah[0] + vektor[0], tengah[1] + vektor[1]] # Implementasi list comprehension
    tengah_baru = tengah + vektor # Implementasi NumPy (asumsi 'tengah' adalah np.array)
    return tengah_baru

# Fungsi utama untuk input dinamis dan translasi
def main():
    # Input pusat lingkaran dan radius dari pengguna
    print("Masukkan pusat lingkaran (dalam format x, y):")
    x, y = map(float, input("Pusat Lingkaran (x, y): ").split(','))
    
    radius = float(input("Masukkan radius lingkaran: "))
    
    # Pusat lingkaran awal
    pusat_lingkaran_awal = np.array([x, y])

    # Input vektor translasi dari pengguna
    print("Masukkan vektor translasi (dalam format dx, dy):")
    dx, dy = map(float, input("Vektor Translasi (dx, dy): ").split(','))
    vektor_translasi = np.array([dx, dy])

    # Translasi pusat lingkaran
    pusat_lingkaran_tertranslasi = translasi(pusat_lingkaran_awal, vektor_translasi)

    # Plot lingkaran sebelum dan sesudah translasi
    plt.figure()

    # Gambar lingkaran awal
    gambar_lingkaran(pusat_lingkaran_awal, radius, warna='b', label='Lingkaran Awal')

    # Gambar lingkaran setelah translasi
    gambar_lingkaran(pusat_lingkaran_tertranslasi, radius, warna='r', label='Lingkaran Setelah Translasi')

    # Menampilkan grid dan setting sumbu
    plt.grid(True)
    
    # Menentukan batas sumbu X dan Y
    # Batas sumbu ditentukan oleh pusat lingkaran yang sudah ditranslasi dan radius.
    plt.xlim(pusat_lingkaran_tertranslasi[0] - radius - 1, pusat_lingkaran_tertranslasi[0] + radius + 1)
    plt.ylim(pusat_lingkaran_tertranslasi[1] - radius - 1, pusat_lingkaran_tertranslasi[1] + radius + 1)
    
    plt.gca().set_aspect('equal', adjustable='box')
    plt.legend()
    plt.title('Translasi 2D Lingkaran')

    # Tampilkan plot
    plt.show()

# Panggil fungsi utama
main()
```
